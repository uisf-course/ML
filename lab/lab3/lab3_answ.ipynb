```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
def generate_dataset(range_start, range_end, num_sample, noise_std=5):
  X = np.linspace(range_start, range_end, num_sample)
  noise = np.random.normal(0,noise_std, num_sample)
  y = 3 * X + 8 + noise
  return X,y
```

```python
X, y = generate_dataset(-10, 10, 50)
```

visualise

```python
plt.scatter(X,y, label='Data points')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
def h_w(X,w) :
  return X * w[1] + w[0]
```

```python
def cost_function(X,y,w) :
  m = len(X)
  return np.sum((y-h_w(X,w))**2) / len(X)
```

```python
w0_vals = np.linspace(-10,20, 200)
w1_vals = np.linspace(-1,5, 200)
j_vals = np.zeros((len(w0_vals),len(w1_vals)))

for i in range(len(w0_vals)) :
  for j in range(len(w1_vals)) :
    w = [w0_vals[i], w1_vals[j]]
    j_vals[i,j] = cost_function(X, y, w)

```

```python
ax = plt.axes(projection='3d')
w0,w1 = np.meshgrid(w0_vals, w1_vals)
ax.plot_surface(w0, w1, j_vals, cmap="viridis")
ax.set_xlabel("w0")
ax.set_ylabel("w1")
```
